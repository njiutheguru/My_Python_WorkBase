```python
page= requests.get("https://forecast.weather.gov/MapClick.php?lat=34.0536&lon=-118.2454")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
seven_day_forecast=soup.find(id="seven-day-forecast")
forecast_items=seven_day_forecast.find_all(class_="tombstone-container")
```

```python
today=forecast_items[0]
```

```python
print(today.prettify())
```

```python
period=today.find(class_='period-name').get_text()
```

```python
print(period)
```

```python
short_desc=today.find(class_='short-desc').get_text()
```

```python
print(short_desc)
```

```python
temp=today.find(class_='temp').get_text()
```

```python
print(temp)
```

```python
img=today.find('img')
desc=img['title']
print(desc)
```

```python
# Using Css selectors in web scraping
```

```python
period_tags=seven_day_forecast.select(".tombstone-containers period.name")
```

```python
period_tags
```

```python
periods=[pt.get_text() for pt in period_tags]
```

```python
periods
```

```python
print(periods)
```

# Loading data to a Pandas Dataframe

```python
import pandas as pd
```

```python
weather =pd.DataFrame({
    'period':periods,
    'short_desc': short_desc,
    'temp': temp,
    'desc': desc
})
```

```python
print(weather)
```

```python
search='(\d+)'
temp_nums=weather['temp'].str.extract(search,expand=False)
weather["temp_num"] =temp_nums.astype('int')
```

```python
weather['temp_num'].mean()
```

```python
is_low=weather['temp'].str.contains('Low')
weather['is_low']=is_low
weather=weather[weather['is_low']]
```
